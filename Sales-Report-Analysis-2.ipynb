# Sales Report Analysis 2

## Import Library

```python
import pandas as pd
import numpy as np
```

## Import Data

```python
df1 = pd.read_excel('report1.xlsx')
df2 = pd.read_excel('report2.xlsx')
df3 = pd.read_excel('report3.xlsx')

display(df1,df2,df3)
```

```python
df2.insert(1, "Continent", ['East Asia', 'Southeast Asia', 'South America', 'Southeast Asia', 'Southeast Asia', 'South America', 'East Asia', 'Europe', 'Central Asia', 'Europe', 'Europe', 'Middle East', 'Southeast Asia'], True) 
df2.head()
```

## Merge Table

```python
df4 = pd.merge(df1, df2, how='right')
df4.head()
```

```python
df5 = pd.merge(df4, df3, how='outer')
df5.head()
```

```python
df5.isnull().any()
```

## Calculate Sales 

```python
df5['Sales(US$)'] = df5['Unit Price (US$)'] * df5['Quantity']
df5.head()
```

## Calculate Delivery Time & Tracking

```python
df5['Delivery_Time'] = df5['Delivery Date'] - df5['Purchase Date']
df5.head()
```

```python
df5['Delivery_Time'] = df5['Delivery_Time'].dt.days.astype('int16')
df5.head()
```

```python
df5['Delivery_Tracking'] = df5.Delivery_Time.apply(lambda x: 'LATE' if x >= 12 else 'ON TRACK')
df5.head()
```

## Calculate Incentive

```python
def conditions(s):
    if s['Sales(US$)'] >= 2000 and s['Delivery_Time'] < 12:
        return s['Sales(US$)']* 0.05
    elif s['Sales(US$)'] >= 5000 and s['Delivery_Time'] < 12:
        return s['Sales(US$)']* 0.1
    elif s['Sales(US$)'] >= 10000 and s['Delivery_Time'] < 12:
        return s['Sales(US$)']* 0.2
    else:
        return 
```

```python
df5['Incentive(US$)'] = df5.apply(conditions, axis=1)
df5.head()
```

## Visualisation

```python
df6 = df5[['Continent', 'Sales(US$)']]
df6.head()
```
